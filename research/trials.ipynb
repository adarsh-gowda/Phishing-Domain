```python
# Specify the target column
target_column = 'phishing'

# Separate features (X) and target (y)
X = df.drop(columns=[target_column])  # Features (everything except the target)
y = df[target_column]  # Target (the column you want to predict)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
X = add_constant(X)
```

```python
# Function to calculate VIF
def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data[vif_data["Feature"] != "const"]  # Exclude constant
```

```python
X
```

```python
calculate_vif(X)
```

```python
# Set VIF threshold
threshold = 5
```

```python
# Batch removal process
def batch_remove_high_vif(X, threshold):
    while True:
        # Calculate VIF
        vif_data = calculate_vif(X)
        
        # Find features with VIF greater than the threshold
        high_vif_features = vif_data[vif_data["VIF"] > threshold]["Feature"]
        
        if len(high_vif_features) == 0:
            # Break if no features have VIF greater than the threshold
            break
        
        # Drop all features with VIF greater than the threshold in one go
        X = X.drop(columns=high_vif_features)
        print(f"Dropping features with VIF > {threshold}: {list(high_vif_features)}")
    
    return X

# Apply the function to your dataset
X_cleaned = batch_remove_high_vif(X, threshold)

print(f"Final set of features after removing those with VIF > {threshold}:")
print(X_cleaned.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from statsmodels.tools.tools import add_constant
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y, test_size=0.2, random_state=42)
```

```python
# Initialize the Logistic Regression model
model = LogisticRegression(max_iter=1000)  # Increase max_iter if needed
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python

# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python

# Print classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python

# Print confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Example of adjusting threshold
probabilities = model.predict_proba(X_test)[:, 1]  # Get probabilities for class 1
y_pred_adjusted = (probabilities >= 0.4).astype(int)  # Adjust threshold to 0.4
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_adjusted)
print(f"Accuracy: {accuracy:.2f}")
```

```python

# Print classification report
print("Classification Report:")
print(classification_report(y_test, y_pred_adjusted))
```

```python
from sklearn.metrics import roc_auc_score, average_precision_score

roc_auc = roc_auc_score(y_test, probabilities)
pr_auc = average_precision_score(y_test, probabilities)

print(f"AUC-ROC: {roc_auc:.2f}")
print(f"AUC-PR: {pr_auc:.2f}")
```
