```python
# Specify the target column
target_column = 'phishing'

# Separate features (X) and target (y)
X = df.drop(columns=[target_column])  # Features (everything except the target)
y = df[target_column]  # Target (the column you want to predict)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
X = add_constant(X)
```

```python
# Create a DataFrame to store VIF values
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns

# Calculate VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
```

```python
# Display the VIF values
print(vif_data)
```

```python
# Function to calculate VIF
def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data[vif_data["Feature"] != "const"]  # Exclude constant
```

```python
# Set VIF threshold
threshold = 5
```

```python
# Batch removal process
def batch_remove_high_vif(X, threshold):
    while True:
        # Calculate VIF
        vif_data = calculate_vif(X)
        
        # Find features with VIF greater than the threshold
        high_vif_features = vif_data[vif_data["VIF"] > threshold]["Feature"]
        
        if len(high_vif_features) == 0:
            # Break if no features have VIF greater than the threshold
            break
        
        # Drop all features with VIF greater than the threshold in one go
        X = X.drop(columns=high_vif_features)
        print(f"Dropping features with VIF > {threshold}: {list(high_vif_features)}")
    
    return X

# Apply the function to your dataset
X_cleaned = batch_remove_high_vif(X, threshold)

print(f"Final set of features after removing those with VIF > {threshold}:")
print(X_cleaned.columns)
```
